## 一.python实现的筛图程序

* ### 起因

最初的线下筛图采用的是拖拽文件夹（多分类）

![](http://zhixuan001.wang/wzx/somepic/001.png)

或者删除一类（二分类）的方法

![](http://zhixuan001.wang/wzx/somepic/002.png)

而且人脸关键点的审核比较复杂

![](http://zhixuan001.wang/wzx/somepic/003.jpg)

一旦有误操作会非常难以处理

#### *缺点：*

1.费时费力，效率低下

2.工序繁琐，极易出错

* ### 需求

一个跨平台的易用的筛图程序，能够提高效率

#### *需要实现的功能：*

1.图形化界面（GUI）

2.读取并显示图片

3.可分部位审核

4.将分好类的图片分到不同的文件夹下

5.误操作的可以回去修改标注

* ### 实现

第一版

![](http://zhixuan001.wang/wzx/somepic/004.png)

![](http://zhixuan001.wang/wzx/somepic/005.png)

* ### 完善

*在使用过程中不断的发现新问题，提出新的需求：*

1.图片的放大缩小的功能

2.根据序号或文件名进行跳转的功能

3.集成读写图片名txt文件的功能

4.显示文件名、进度、各分类占比这些标注信息的功能

5.中途退出下次可续做功能

6.将未标注的图片取出，供再分配或者任务交接的功能

7.增加键盘按键标注的功能

8.防止用户各种误操作

<br />

*为了适配更多的项目，还加入了这些功能：*

1.支持更多的图片格式

2.支持处理无法读取的图片

3.支持旋转图片

4.支持自定义分类按钮

5.简化出一个简化版本，去掉了人脸项目独有的功能，界面更简洁

<br />

现在已经更新到第四版

简化版的界面
